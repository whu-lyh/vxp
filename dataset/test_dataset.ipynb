```python
import random 
random.seed(0)
```

```python
import matplotlib.pyplot as plt

def plot_submap_image(submap, img):
    # fig, axs = plt.subplots(1, 2)
    fig = plt.figure(figsize=(15, 8))
    ax_submap = fig.add_subplot(1, 2, 1, projection='3d', )
    ax_submap.scatter(submap[:, 0], submap[:, 1], submap[:, 2], s=0.1)
    ax_submap.set_xlabel('X')
    ax_submap.set_ylabel('Y')
    ax_submap.set_zlabel('Z')
    ax_submap.set_xlim(0, 50)
    ax_submap.set_ylim(-25, 25)
    ax_submap.view_init(elev=-150, azim=30)
    
    ax_img = fig.add_subplot(1, 2, 2)
    ax_img.imshow(img)
    plt.show()
```

```python
idx = random.choice(range(len(dataset_place_recognition)))
print(idx)
img_anchor, submap_anchor, img_pos, submap_pos, img_neg, submap_neg = dataset_place_recognition[idx]
plot_submap_image(submap=submap_anchor, img=img_anchor)
plot_submap_image(submap=submap_pos, img=img_pos)
plot_submap_image(submap=submap_neg, img=img_neg)
```
